```python
from langchain_openai import AzureChatOpenAI
from langchain_core.messages import AIMessage,SystemMessage,HumanMessage
from dotenv import load_dotenv
load_dotenv()

llm = AzureChatOpenAI(
    azure_endpoint="",
    azure_deployment="",
    api_version="",
    model=''
)
```

```python
from langchain.memory import ConversationBufferMemory

#stores entire chat history in text
memory = ConversationBufferMemory(return_messages=True)
memory.save_context({"input": "Hi, I am John"}, {"output": "Nice to meet you John!"})
print(memory.load_memory_variables({}))
```

```python
from langchain.memory import ConversationBufferWindowMemory

#Stores only the last N messages.
memory = ConversationBufferWindowMemory(k=3, return_messages=True)

memory.save_context({"input": "Message 1"}, {"output": "OK"})
memory.save_context({"input": "Message 2"}, {"output": "OK"})
memory.save_context({"input": "Message 3"}, {"output": "OK"})
memory.save_context({"input": "Message 4"}, {"output": "OK"})
memory.save_context({"input": "Message 5"}, {"output": "OK"})
memory.save_context({"input": "Message 6"}, {"output": "OK"})

print(memory.load_memory_variables({}))
print(memory)
```

```python
from langchain.memory import ConversationSummaryMemory
#Stores long conversations by summarizing automatically. #use llmchain module and llm call to summarize it 
memory = ConversationSummaryMemory(llm=llm)

memory.save_context({"input": "Hello I am Mark"}, {"output": "Nice to meet you Mark"})
memory.save_context({"input": "I work at Google"}, {"output": "Great!"})

print(memory.load_memory_variables({}))
```
